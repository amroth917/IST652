```
#Question 1
#Which customer has the largest purchase on black friday? What's their gender and age?

#bf['Gender'].value_counts()
#bf['Purchase'].max()
#bf['Age'].value_counts()

bfmax = bf.loc[bf['Purchase'].idxmax()]

print("The customer with the largest purchase on Black Friday is", bfmax[2],"and",bfmax[3], 
      ", spending a total of ${:,.2f}".format(bfmax[9]))

bf.to_csv(r'Path where you want to store the exported CSV file\File Name.csv')

bfmax_export = bfmax.to_csv (r'//BlackFridayMaxPurchase.csv', index= None, header=True)
```

```
#Question 2
#What was the average purchase within each age group?

#df2 = {n: rows[1] for n, rows in enumerate(bf.groupby('Gender'))}
#print(df2[1])

bfage = bf.groupby(['Age'], as_index=False).mean().groupby('Age').mean()

#print("On average, people within", bfage.index[0],"age group spent:",bfage['Purchase'][0])

count = 0
for line in bfage:
    print ("On average, people within", bfage.index[count],"age group spent ${:,.2f} on Black Friday.".format(bfage['Purchase'][count]))
    count += 1
    
    
bfage_export = bfage.to_csv (r'//BlackFridayAgeAvg.csv', index= None, header=True)

```
